```python
import numpy as np
import tensorflow as tf 
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense
from tensorflow.keras import Input
from tensorflow.keras.layers import concatenate
import matplotlib.pyplot as plt
from tensorflow.keras.utils import plot_model
from tensorflow import keras
from tensorflow.keras import layers
```

```python
def f_classifieur(v):
    y = []
    for x in v:
        if (x>-0.5 and x<0.5):
            y.append(1)
        else:
            y.append(0)
    return np.array(y)
```

```python
seed = 7
np.random.seed(seed)

N = 1000
N_val = 200
X_train = np.linspace(0.0,1.0,N)
y_train = np.sin(2*np.pi*X_train)
X_train2 = np.linspace(-1.0,1.0,N)
y_model = f_classifieur(X_train2)

X_val = np.random.rand(N_val)
X_val = np.sort(X_val)
y_val = np.sin(2*np.pi * X_val)
y_val_model = f_classifieur(X_val)
```

```python

# Define Sequential model with 3 layers
model1 = keras.Sequential(
    [
        layers.Dense(64, input_dim = 1,activation="relu", name="layer1"),
        layers.Dense(64, activation="relu", name="layer2"),
        layers.Dense(64, activation="relu", name="layer3"),
        layers.Dense(1, activation="sigmoid", name="layer4"),
    ]
)
```

```python
#compile the model and compute the parameters of the model
model1.compile(optimizer='rmsprop',loss='mse', loss_weights=[1,1],metrics='mse')
model1.fit(X_train2,y_model,epochs=50,batch_size=32,validation_data=(X_val,y_val_model))
```

```python
y_pred = model1.predict(X_train2)
```

```python
plt.plot(X_train2,y_model,'b')
plt.plot(X_train2,y_pred,'r')
plt.plot(X_val,y_val_model,'g')
```

```python

# Define Sequential model with 3 layers
model2 = keras.Sequential(
    [
        layers.Dense(64, input_dim = 1,activation="sigmoid", name="layer1"),
        layers.Dense(64, activation="sigmoid", name="layer2"),
        layers.Dense(64, activation="sigmoid", name="layer3"),
        layers.Dense(1, name="layer4"),
    ]
)
```

```python
#compile the model and compute the parameters of the model
model2.compile(optimizer='rmsprop',loss='mse', loss_weights=[1,1],metrics='mse')
model2.fit(X_train,y_train,epochs=300,batch_size=32,validation_data=(X_val,y_val))
```

```python
y_pred1 = model2.predict(X_train)
```

```python
y_pred2 = model1.predict(y_pred1)
```

```python
plt.plot(X_train2,np.sin(2*np.pi * X_train2),'b')
plt.plot(X_train2,y_pred2*y_pred1,'r')
```

```python
def f_final(y1,y2):
    res = np.zeros(np.size(y1))
    for i in range(np.size(y1)):
        if y2[i]>0:
            res[i] = y1[i]
        elif y1[i]>0:
            res[i] = 0.5
        else:
            res[i] = -0.5
    return res
```

```python
plt.plot(X_train,np.sin(2*np.pi * X_train),'b')
plt.plot(X_train,f_final(y_pred1,y_pred2),'r')
# plt.plot(X_train,y_pred1,'g')
```

```python
plt.plot(X_train,np.sin(2*np.pi * X_train),'b')
plt.plot(X_train,f_final(y_pred1,y_pred2),'r')
plt.plot(X_train,y_pred1,'g')
```
