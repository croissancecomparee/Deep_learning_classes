```python
import numpy as np
import tensorflow as tf 
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense
from tensorflow.keras import Input
from tensorflow.keras.layers import concatenate
import matplotlib.pyplot as plt
from tensorflow.keras.utils import plot_model
from tensorflow import keras
from tensorflow.keras import layers
```

```python
seed = 7
np.random.seed(seed)

N = 1000
X_train = np.linspace(0.0,1.0,N)
y_train = np.sin(2*np.pi*X_train)

X_val = np.random.rand(100)
X_val = np.sort(X_val)
y_val = np.sin(2*np.pi * X_val)
```

```python
plt.plot(X_val , y_val , 'r-')

plt.plot(X_train , y_train , 'b-')
plt.title('Training and validation Data')
```

```python
from tensorflow import keras
from tensorflow.keras import layers

# Define Sequential model with 3 layers
model = keras.Sequential(
    [
        layers.Dense(32, input_dim = 1,activation="sigmoid", name="layer1"),
        layers.Dense(16, activation="sigmoid", name="layer2"),
        layers.Dense(1, name="layer3"),
    ]
)
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

def custom_loss():
    with tf.GradientTape(persistent=True) as tape:
        X_train_1 = tf.cast(X_train, tf.float32)  # Ensure X_train is float32
        y = model(X_train_1)
        pi = tf.constant(np.pi, dtype=tf.float32)
        loss = tf.reduce_mean(tf.square(tf.sin(2 * pi * X_train_1) - y))
        grad = tape.gradient(loss, model.trainable_variables)
        optimizer.apply_gradients(zip(grad, model.trainable_variables))
    
    return loss
```

```python
X_train = X_train.reshape([-1,1])
```

```python
X_train = tf.Variable(X_train)
```

```python
losses = []
for i in range(5000):
    loss = custom_loss()
    losses.append(loss)
    print("loss", loss.numpy())
```

```python
plt.plot(losses)
```

# REPREND A ZERO

```python
def custom_loss(y_trace, y_pred):
    loss = tf.reduce_mean(tf.square(y_pred - y_trace)) + tf.reduce_mean(y_pred - 0.5) + tf.reduce_mean(y_pred - 0.5)
    return loss
```

```python

# Define Sequential model with 3 layers
model = keras.Sequential(
    [
        layers.Dense(32, input_dim = 1,activation="sigmoid", name="layer1"),
        layers.Dense(16, activation="sigmoid", name="layer2"),
        layers.Dense(1, name="layer3"),
    ]
)
```

```python
#compile the model and compute the parameters of the model
model.compile(optimizer='rmsprop',loss=custom_loss, loss_weights=[1,1],metrics='mse')
model.fit(X_train,y_train,epochs=300,batch_size=32,validation_data=(X_val,y_val))
```

```python
y_pred = model.predict(X_train)
```

```python
plt.plot(X_train,y_pred,'b')
plt.plot(X_val,y_val,'r')
plt.title('Training and validation data for output 1')
```
